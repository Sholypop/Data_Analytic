```python
import requests
import bs4
```

```python
#create a variable for the url i intend to crapp from and make a request
url_result = requests.get('https://www.worldometers.info/geography/7-continents/north-america/')
```

```python
type(url_result)
```

```python
url_result.text
```

```python
#create a soup variable using beautiful soup
soup = bs4.BeautifulSoup(url_result.text, 'lxml')
soup
```

```python
#import other library needed to perform furher data manipulation
import pandas as pd
import numpy as np
import io
```

```python
#inspect desired table i want to crap and use the information to select and extract what i need from the website
table1 = soup.find('table', class_='table table-hover table-condensed')
```

```python
#creating a for loop to extract and compile the column name
# this will be stored in a variable called headers
headers = []
for i in table1.find_all('th'):
    column_title= i.text
    headers.append(column_title)
```

```python
headers
```

```python
headers[0]= 'index'
```

```python
headers
```

```python
mydata = pd.DataFrame(columns = headers)
mydata
```

```python
#creating a loop to fill mydata column

for j in table1.find_all('tr')[1:]:  #important to start from loc 1, as data at loc 0 has been extracted already into headers
    row_data = j.find_all('td')
    row= [i.text for i in row_data]
    length = len(mydata)
    mydata.loc[length] = row
```

```python
mydata
```

```python
#clean data frame and dropp unwanted columns
mydata.drop('index', inplace=True, axis=1)
```

```python
mydata
```

```python
#save data into csv and get it ready to be used for further data analysis work or visualisation. 
mydata.to_csv('webscrappeddata_project.csv')
```
